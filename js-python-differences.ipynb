const reverseIt = () => {
  let string = "a man, a plan, a canal, frenemies!";

  let reverse = "";

  for (let i=0; i < string.length; i++) {
    reverse += string[string.length - (i+1)];
  };

  alert(reverse);
};

````python
def reverse_it(str_text):

    reverse = []

    str_length = len(str_text)

    for i in range(str_length):
        reverse.append(str_text[str_length-1-i])

    reversed_string = ''.join(reverse)

    return reversed_string

print(f"{reverse_it("a man, a plan, a canal, frenemies!")=}")
````
